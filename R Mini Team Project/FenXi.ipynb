```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
YS = pd.read_excel('YS.xlsx')
BS = pd.read_excel('BS.xlsx')
GR = pd.read_excel('GR.xlsx')
JJ = pd.read_excel('JJ.xlsx')
TY = pd.read_excel('TY.xlsx')
```

```python
# 대분류 검색 건수
YSsum = YS.groupby('대분류')['외지인검색수'].sum()
BSsum = BS.groupby('대분류')['외지인검색수'].sum()
GRsum = GR.groupby('대분류')['외지인검색수'].sum()
JJsum = JJ.groupby('대분류')['외지인검색수'].sum()
TYsum = TY.groupby('대분류')['외지인검색수'].sum()
```

```python
# 대분류 검색 건수 시각화

fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(2, 3, 1, title='여수')
ax2 = fig.add_subplot(2, 3, 2, title='부산')
ax3 = fig.add_subplot(2, 3, 3, title='강릉')
ax4 = fig.add_subplot(2, 3, 4, title='제주도')
ax5 = fig.add_subplot(2, 3, 5, title='통영')


ax1.pie(YSsum, labels = YSsum.index, colors=['#ffc952', '#2EC4B6', '#75D701'], autopct='%1.1f%%',
       shadow=True, startangle=-90, explode=(0.05,0,0))


ax2.pie(BSsum, labels = BSsum.index, colors=['#ffc952', '#2EC4B6', '#75D701'], autopct='%1.1f%%',
       shadow=True, startangle=-90, explode=(0.05,0,0))


ax3.pie(GRsum, labels = GRsum.index, colors=['#ffc952', '#2EC4B6', '#75D701'], autopct='%1.1f%%',
       shadow=True, startangle=-90, explode=(0.05,0,0))


ax4.pie(JJsum, labels = JJsum.index, colors=['#ffc952', '#2EC4B6', '#75D701'], autopct='%1.1f%%',
       shadow=True, startangle=-90, explode=(0.05,0,0))


ax5.pie(TYsum, labels = TYsum.index, colors=['#ffc952', '#2EC4B6', '#75D701'], autopct='%1.1f%%',
       shadow=True, startangle=-90, explode=(0.05,0,0))

ax3.legend(GRsum.index, loc=4, bbox_to_anchor=(0.85,-0.9), fontsize=20)
plt.show()
```

```python
# 소분류 자연

YS_Zsum = YS.loc[YS.대분류 == '자연',:].groupby('소분류')['외지인검색수'].sum()
BS_Zsum = BS.loc[BS.대분류 == '자연',:].groupby('소분류')['외지인검색수'].sum()
GR_Zsum = GR.loc[GR.대분류 == '자연',:].groupby('소분류')['외지인검색수'].sum()
JJ_Zsum = JJ.loc[JJ.대분류 == '자연',:].groupby('소분류')['외지인검색수'].sum()
TY_Zsum = TY.loc[TY.대분류 == '자연',:].groupby('소분류')['외지인검색수'].sum()
```

```python
# 소분류 검색 건수 시각화

fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(2, 3, 1, title='여수',)
ax2 = fig.add_subplot(2, 3, 2, title='부산')
ax3 = fig.add_subplot(2, 3, 3, title='강릉')
ax4 = fig.add_subplot(2, 3, 4, title='제주도')
ax5 = fig.add_subplot(2, 3, 5, title='통영')

color = ['#30A9DE', '#75D701', '#FADAD8', 'Violet', 'Yellow','Orange','gold']

ax1.pie(YS_Zsum, colors=color, 
        autopct='%1.1f%%', shadow=True, startangle=-90, explode=(0.05,0))


ax2.pie(BS_Zsum, colors=color, 
        autopct='%1.1f%%', shadow=True, startangle=-90, explode=(0.05,0))


ax3.pie(GR_Zsum, colors=color, 
        autopct='%1.1f%%', shadow=True, startangle=-90, explode=(0.05,0,0))


ax4.pie(JJ_Zsum, colors=color, 
        autopct='%1.1f%%', shadow=True, startangle=-90, explode=(0.05,0,0))


ax5.pie(TY_Zsum, colors=color, 
        autopct='%1.1f%%', shadow=True, startangle=-90, explode=(0.05,0))

ax3.legend(GR_Zsum.index, loc=4, bbox_to_anchor=(0.83,-0.89), fontsize=20)
plt.show()
```
